# Task2-> Displaying IMDB’s Top rated 100 movies’ data

```python
page1=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
page1
```

```python
page2=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
page2
```

The data is present in 2 pages,so used two different beautiful soup objects.

```python
soup1=BeautifulSoup(page1.content,'html.parser')
soup1
```

```python
soup2=BeautifulSoup(page2.content,'html.parser')
soup2
```

```python
movies=soup1.find_all("h3",class_="lister-item-header")
```

```python
movies_rest=soup2.find_all("h3",class_="lister-item-header")
movies_rest[49].text
```

```python
len(movies)
```

```python
len(movies_rest)
```

```python
#placing the first 50 movies data into list.As the data contains \n,splitting the entries based on \n.
movies_50=[movies[i].text.split("\n") for i in range(len(movies))]
movies_50
```

```python
#converting the rest of 50 into list.
movies_100=[movies_rest[i].text.split("\n")  for i in range(len(movies_rest))]
movies_100
```

```python
#combining the both movies into a single list.
top100_movies=movies_50+movies_100
```

```python
top100_movies
```

```python
#converting the list to dataframe.
movies_df=pd.DataFrame(top100_movies)
```

```python
movies_df
```

```python
#dropping 0 and 4 columns as they dont possess any values.
movies_df.drop([0],axis=1,inplace=True)
movies_df.drop([4],axis=1,inplace=True)
```

```python
#renaming the columns for better understanding.
movies_df.rename(columns={1:'Rank',2:'Movie Name',3:'Release Year'},inplace=True)
movies_df
```

```python
#removing the braces in release year column.
movies_df['Release Year']=movies_df['Release Year'].str.replace("(","")
movies_df['Release Year']=movies_df['Release Year'].str.replace(")","")
```

```python
movies_df
```

Adding the ratings column for the dataframe.


```python
#creating list for ratings of all the 100 movies
mvis_rating50=soup1.find_all("div",class_='inline-block ratings-imdb-rating')
mvis_rating100=soup2.find_all("div",class_='inline-block ratings-imdb-rating')
mvis_ratings=mvis_rating50+mvis_rating100
mvis_ratings=[mvis_ratings[i].text.strip("\n")  for i in range(len(mvis_ratings))]
```

```python
mvis_ratings
```

Adding ratings columns to existing dataframe.

```python
movies_df['Rating']=mvis_ratings
```

```python
#final dataframe.
movies_df
```

```python
#saving the results to csvfile.
movies_df.to_csv('IMDB Top 100 Movies.csv')
```
